```python
from binarize import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
images = list(map(imageio.imread, ['gradient.png', 'lena.bmp', 'newspaper.jpg', 'camera.jpg']))
```

```python
def show(func, downsampling=1):
    size = 10
    axes = plt.subplots(len(images), 1, figsize=(size, size * len(images)))[1]
    for a, i in zip(np.array(axes).flatten(), images):
        a.imshow(func(i[::downsampling, ::downsampling]), cmap='gray')
        a.set_axis_off()
```

```python
show(lambda x: x)
```
